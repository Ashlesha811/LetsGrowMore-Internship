###  Exploratory Data Analysis on Dataset - Terrorism

```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
terrorism = pd.read_csv('terrorism.csv',encoding='ISO-8859-1')
```

```python
terrorism
```

```python
terrorism.columns
```

```python
for i in terrorism.columns:
    print(i)
```

```python
terrorism=terrorism.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','region_txt':'Region',
                             'provstate':'State','city':'City','attacktype1_txt':'Attack_type','targtype1_txt':'Target_type',
                             'targsubtype1_txt':'Target_subtype','gname':'Gang_name','weaptype1_txt':'Weapon_type'})
```

```python
terrorism.isnull().sum()
```

```python
terror=terrorism[['Year','Month','Day','Country','Region','City','latitude','longitude','State','Attack_type','nkill',
               'nwound','target1','summary','Gang_name','Target_type','Weapon_type','motive']]
```

```python
terror.isnull().sum()
```

```python
terror.info()
```

```python
terror['City']=terror['City'].fillna('Unknown')
terror['State']=terror['State'].fillna('Unknown')
```

```python
terror['target1'] = terror['target1'].fillna('others')
```

```python
print("Country with the most attacks:",terror['Country'].value_counts().idxmax())
print("City with the most attacks:",terror['City'].value_counts().index[1]) 
print("Region with the most attacks:",terror['Region'].value_counts().idxmax())
print("Year with the most attacks:",terror['Year'].value_counts().idxmax())
print("Month with the most attacks:",terror['Month'].value_counts().idxmax())
print("Group with the most attacks:",terror['Gang_name'].value_counts().index[1])
print("Most Attack Types:",terror['Attack_type'].value_counts().idxmax())
```

```python
plt.figure(figsize=(25,10))
sns.countplot(data=terror,x='Year')
plt.title('Number Of Terrorist Activities Each Year')
```

```python
plt.figure(figsize=(10,7))
sns.countplot(data=terror,x='Month')
```
